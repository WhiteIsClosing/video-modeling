```python
import sys 
import numpy
import numpy.random
import theano
import theano.tensor as T
from theano.tensor.shared_randomstreams import RandomStreams
from scipy import misc
from time import clock
numpy_rng  = numpy.random.RandomState(1)
theano_rng = RandomStreams(1)
```

```python
from hyper_params import *
sys.path.append(project_path)

from utils.load import *
from utils.plot import *
from utils.log import *

from grammar_cells.pgp3layer import *
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

## Load Data

```python
# LOAD DATA
features_numpy = numpy.load(data_path + 'translation.npy').astype('float32')
features_train_numpy = features_numpy[:numseqs_train, :seq_len*frame_dim]
features_test_numpy =\
features_numpy[numseqs_train:numseqs_train+numseqs_test,
                    :seq_len*frame_dim]

true_frames_train = features_train_numpy
true_frames_test = features_test_numpy

# PREPROCESS
data_mean = true_frames_train.mean()
data_std = true_frames_train.std()
true_frames_train -= data_mean
true_frames_train /= data_std 
true_frames_test -= data_mean
true_frames_test /= data_std
```

## Initialize Model

```python
model = Pgp3layer(numvis=frame_dim,
                    numnote=0,
                    numfac=numfac,
                    numvel=numvel,
                    numvelfac=numvelfac,
                    numacc=numacc,
                    numaccfac=numaccfac,
                    numjerk=numjerk,
                    seq_len_to_train=seq_len_to_train,
                    seq_len_to_predict=seq_len_to_predict,
                    output_type='real',
                    vis_corruption_type='zeromask',
                    vis_corruption_level=0.0,
                    numpy_rng=numpy_rng,
                    theano_rng=theano_rng)
```

## Load Model

```python
model.load(models_path + 'model.npy')
```

## Plot Filter Pairs

```python
grid_left = plotPatchGrid(model.wxf_left.get_value().T.flatten(), image_shape, 10, margin=2, margin_val=1e0)
grid_right = plotPatchGrid(model.wxf_right.get_value().T.flatten(), image_shape, 10, margin=2, margin_val=1e0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20)) # figsize for scaling

fig1 = ax1.imshow(grid_left, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax1.axis('off')
ax1.set_title('wdf_left')
fig1.set_interpolation('nearest')

ax2.imshow(grid_right, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax2.axis('off')
ax2.set_title('wdf_right')

plt.show()
```

## Predict Sequences

```python
pred_frames_train = model.predict(true_frames_train, 20)
pred_frames_test = model.predict(true_frames_test, 20)
```

## Plot Predictions

```python
maxplot = 100 # maxial sequence length to plot

# train
grid_true_frames_train = plotPatchGrid(features_train_numpy, image_shape, 20, margin=2, margin_val=5.e0, max_num=maxplot)
grid_pred_frames_train = plotPatchGrid(pred_frames_train, image_shape, 20, margin=2, margin_val=5.e0, max_num=maxplot)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

ax1.imshow(grid_true_frames_train, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax1.axis('off')
ax1.set_title('true frames (trainset)')

ax2.imshow(grid_pred_frames_train, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax2.axis('off')
ax2.set_title('predicted frames (trainset)')

# test
grid_true_frames_test = plotPatchGrid(features_test_numpy, image_shape, 20, margin=2, margin_val=5.e0, max_num=maxplot)
grid_pred_frames_test = plotPatchGrid(pred_frames_test, image_shape, 20, margin=2, margin_val=5.e0, max_num=maxplot)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

ax1.imshow(grid_true_frames_test, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax1.axis('off')
ax1.set_title('true frames (testset)')

ax2.imshow(grid_pred_frames_test, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax2.axis('off')
ax2.set_title('predicted frames (testset)')
```

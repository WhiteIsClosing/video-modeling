## Plot Filter Pairs

```python
grid_left = plotPatchGrid(model.wfd_left.get_value().flatten(), image_shape, 10, margin=2, margin_val=1.e0)
grid_right = plotPatchGrid(model.wfd_right.get_value().flatten(), image_shape, 10, margin=2, margin_val=1.e0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20)) # figsize for scaling

fig1 = ax1.imshow(grid_left, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax1.axis('off')
ax1.set_title('wdf_left')
fig1.set_interpolation('nearest')

ax2.imshow(grid_right, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax2.axis('off')
ax2.set_title('wdf_right')

plt.show()
```

## Plot Predictions

```python
maxplot = 100 # maxial sequence length to plot

grid_true_frames_train = plotPatchGrid(features_train_numpy, image_shape, 20, margin=2, margin_val=1.e0, max_num=maxplot)
grid_pred_frames_train = plotPatchGrid(pred_frames_train, image_shape, 20, margin=2, margin_val=1.e0, max_num=maxplot)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

ax1.imshow(grid_true_frames_train, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax1.axis('off')
ax1.set_title('true frames')

ax2.imshow(grid_pred_frames_train, cmap = matplotlib.cm.Greys_r, interpolation ='nearest')
ax2.axis('off')
ax2.set_title('predicted frames')
```
